```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r"transformer-voltage.csv",
                delimiter = ',',
                index_col = 'Sample')
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# z-critical for two-tailed test
alpha = 0.05
z_critical = stats.norm.ppf(1-alpha/2)
z_critical
```

```python
# confidence interval 
pop_std = 5

ci_lower, ci_upper = stats.norm.interval(
    confidence = 0.95,
    loc = df['Voltage'].mean(),
    scale = pop_std/np.sqrt(df['Voltage'].count())
)

print(f"CI:[{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
# voltage distribution
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

# x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color="#2133D6",
        linewidth=2.9)

# histogram
plt.hist(df['Voltage'],
        bins = 5,
        density = True,
        alpha = 0.8,
        edgecolor = '#2133D6')

plt.title('Voltage Distribution',
            font = 'Cambria',
            fontsize = 12)

plt.xlabel('Voltage (V)')
plt.ylabel('Probability Density')

# confidence interval
plt.axvline(x = ci_lower,
            color = "#00010A",
            linestyle = '--')

plt.axvline(x = ci_upper,
            color = "#03030A",
            linestyle = '--')

plt.show()
```
