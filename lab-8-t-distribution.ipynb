```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r"transformer-voltage-10-sample.csv",
                delimiter = ',',
                index_col = ['Sample'])
df
```

```python
df.info()
```

```python
df.describe()
```

```python
# z-critical for two-tailed test
alpha = 0.05
n = df['Voltage'].count()

t_critical = stats.t.ppf(1-alpha/2,n-1)
t_critical
```

```python
# confidence interval 
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

ci_lower, ci_upper = stats.t.interval(
    confidence = 0.95,
    df = n-1,
    loc = df['Voltage'].mean(),
    scale = std/np.sqrt(n)
)

print(f"CI:[{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
# voltage distribution
mean = df['Voltage'].mean()
std = df['Voltage'].std(ddof=1)

# x-axis
x_min = df['Voltage'].min()
x_max = df['Voltage'].max()

x = np.linspace(x_min, x_max, 100)
p = stats.norm.pdf(x, mean, std)

plt.plot(x, p,
        color="#B221D6",
        linewidth=2.9)

# histogram
plt.hist(df['Voltage'],
        bins = 5,
        density = True,
        alpha = 0.8,
        color = "#CF57FF",
        edgecolor = "#570F79")

plt.title('Voltage Distribution',
            font = 'Cambria',
            fontsize = 12)

plt.xlabel('Voltage (V)')
plt.ylabel('Probability Density')

# confidence interval
plt.axvline(x = ci_lower,
            color = "#00010A",
            linestyle = '--')

plt.axvline(x = ci_upper,
            color = "#03030A",
            linestyle = '--')

plt.show()
```

## Shoe Inventory

```python
df = pd.read_csv(r"shoe-inventory.csv",
                delimiter = ',',
)
df
```

```python
# rows and columns
df.shape
```

```python
mean_size_6 = df.iloc[0,1:13].mean()
mean_size_6
```

```python
std_size_6 = df.iloc[0,1:13].std()
std_size_6
```

```python
# size 6 95% CI
n = df.shape[1]-1 #returns 13 in (17,13)

lower_6, upper_6 = stats.t.interval(
    confidence = 0.95,
    df = n-1,
    loc = mean_size_6,
    scale = std_size_6/np.sqrt(n)
)

print(f"CI_6:[{lower_6:.2f}, {upper_6:.2f}]")
```

```python
# 95% Confidence Level 
mean = df.iloc[0:17,1:13].mean(axis = 1)
std = df.iloc[0:17,1:13].std(axis = 1)

df['ci_lower'], df['ci_upper'] = stats.t.interval(
    confidence = 0.95,
    df = n-1,
    loc = mean,
    scale = std/np.sqrt(n)
)

print(f"CI_6:[{ci_lower:.2f}, {ci_upper:.2f}]")
```

```python
df['ci_lower'] = df['ci_lower'].fillna(0)
df['ci_upper'] = df['ci_upper'].fillna(0)
```

```python
df['ci_lower'] = df['ci_lower'].fillna(0).astype(int).round()
df['ci_upper'] = df['ci_upper'].fillna(0).astype(int).round()
```
