```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

## Z-Score

```python
# creating a dataframe
df = pd.DataFrame([1,2,2,3,3,3,4,4,5],
                  columns=['Dataset'])
df
```

```python
df.describe()
```

```python
#Z-Scores
df['Z-score'] = stats.zscore(df['Dataset'],ddof=1)
df
```

```python
df['Z-score'].mean()
```

```python
df['Z-score'].std(ddof=1)
```

## Histogram

```python
plt.hist(df['Current'], bins = 5,
            alpha = 0.8,
            colors = 'light blue',
            edgecolor = 'Black')

plt.title('Histogram of Current Values',
          font = 'Palatino Linotype',
          fontsize = 12)

plt.xlabel('Current (mA)')
plt.ylabel('Frequency')

# Normal Distribution
p = stats.norm.pdf()
plt.show()
```

## Standard Normal Distribution N(0,1)

```python
# Standard Normal Distribution N(0,1)
# x - axis
x_max = df['Z-score'].max()
X_min = df['Z-score'].min
x = np.linspace(x_min,x_max,100)

p = stats.norm.df(x,df['Z-score'].mean,df['Z-score'].std(ddof = 1))

plt.plot(x, p,
         color = 'Blue',
         linewidth = 2)
```
