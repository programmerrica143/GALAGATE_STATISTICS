```python
#import packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
```

```python
df = pd.read_csv(r'current-power.csv',
                    delimiter = ",")
df
```

```python
df.describe()
```

```python
df.cov()
```

```python
x = df['Current']
y = df['Power']

plt.figure()
plt.scatter(x,y)
plt.title('Device Current vs Power Response')
plt.ylabel('Power (W)')
plt.xlabel('Current (A)')
plt.show
```

```python
df.corr()
```

```python
# regression model
slope,intercept,r_value,p_value,std_err = stats.linregress(x, y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize = (7, 4))
plt.scatter(x, y,
            color = "#9823AF",
            alpha = 0.6,
            label= 'Data Points')

# line plot
plt.plot(x, line,
color= "#E214C0",
label = f'Regression line (r = {r_value:.3f})')

plt.title('Device Current vs Power Reponse')
plt.ylabel('Power (W)')
plt.xlabel('current (A)')
plt.legend()
plt.show
```

## Air Quality

```python
# load dataset
air = pd.read_csv(r"air-quality.csv",   
                delimiter = ",")
air
```

## NOx-Humidity

```python
# regression model
x = air['NOx']
y = air['Relative_Humidity']
slope,intercept,r_value,p_value,std_err = stats.linregress(x, y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize = (7, 4))
plt.scatter(x, y,
            color = "#E693F7",
            alpha = 0.2,
            label= 'Data Points')

# line plot
plt.plot(x, line,
color= "#E214C0",
label = f'Regression line (r = {r_value:.3f})')

plt.title('NOx vs Humidity Correlation')
plt.ylabel('Relative Humidity(%)')
plt.xlabel('NOx (ug/cu.m)')
plt.legend()
plt.show
```

## NO2-Humidity

```python
# regression line
x = air['NO2']
y = air['Relative_Humidity']
slope, intercept, r_value, p_value, std_err = stats.linregress(x, y)
line = slope * x + intercept

# scatter plot
plt.figure(figsize=(7, 4))
plt.scatter(x, y,
            color = "#E693F7",
            alpha = 0.2,
            label = 'Data Points')

# line plot
plt.plot(x, line,
         color = "#E214C0",
         label = f'Regression line (r = {r_value:.3f})')

plt.title('NO2 vs Humidity Correlation')
plt.ylabel('Relative Humidity (%)')
plt.xlabel('NO2 (ug/cu.m)')
plt.legend()
plt.show
```
